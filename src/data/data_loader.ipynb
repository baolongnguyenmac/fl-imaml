# Normal data

```python
import matplotlib.pyplot as plt
from data_loader import get_loader
```

```python
# emnist
labels = [f'{i}' for i in range(10)] + [f'B{chr(i)}' for i in range(65, 65+26)] + [f's{chr(i)}' for i in range(97, 97+26)]
fig, axs = plt.subplots(nrows=2, ncols=16, figsize=(35,5))
count = 0
fig.subplots_adjust(hspace=0.2)

loader = get_loader('../../data/emnist/client_test_5/99_q.json', split_supp_query=False, shuffle=False)
for batch in loader:
    for x, y in zip(batch[0], batch[1]):
        axs.flat[count].imshow(x.reshape(28,28).T)
        axs.flat[count].set_title(labels[int(y)])
        count += 1
    break
```
